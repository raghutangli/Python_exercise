### 1. What does RGBA stand for?

        RGBA stands for Red Green Blue Alpha. It's a color model that is used to represent colors in digital images and 
        computer graphics. The first three values (Red, Green, Blue) specify the color of the pixel, while the fourth value 
        (Alpha) specifies its transparency or opacity. The Alpha value is usually represented as a number between 0 and 1, 
        where 0 represents fully transparent and 1 represents fully opaque.



### 2. From the Pillow module, how do you get the RGBA value of any images?

        To get the RGBA value of any image using the Pillow module in Python, we can use the getpixel() method of the
        Image object.

### 5. What method would you call to get Image object for a 100×100 image, excluding the lower-left quarter of it?

```python
from PIL import Image
image = Image.open('BFC.jpg') 
cropped_image = image.crop((150, 200, 800, 400))
cropped_image
```
